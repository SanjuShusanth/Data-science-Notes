## OOP'S PROJECT

1. Student Management: A simple project to display, search, add and delete new students.

Prerequisite: Classes and objects in python 

Problem Statement: Write a program to build a simple Student Management System using Python which can perform the following operations:

1. Accept - This method takes details from the user like name, roll number, and marks for two different subjects.
2. Display - This method displays the details of every student.
3. Search - Â This method searches for a particular student from the list of students. This method will ask the user for roll number and then search according to the roll number
4. Delete - This method deletes the record of a particular student with a matching roll number.
5. Update - This method updates the roll number of the student. This method will ask for the old roll number and new roll number. It will replace the old roll number with a new roll number.


```python
#### step 1 : Creating a class called student_details with the constructor add name, roll_no, social_mark, science_mark.


class Student_Details:
    
    def __init__(self, name, roll_no, social_mark, science_mark):
        
        self.name = name
        self.roll_no = roll_no
        self.social_mark = social_mark
        self.science_mark = science_mark
        
#### Step 2 : creating a display_details inside the class student_details.

    def display_details(self):
        print(f"Name - {self.name}")
        print(f"Roll_no - {self.roll_no}")
        print(f"Social_mark - {self.social_mark}")
        print(f"Science_mark - {self.science_mark}")
        print()
        
#### Step 3 : Implementing another class called StudentManagement in which we add, display, search, delete and update the students details

class StudentManagement:
    
    def __init__(self):
        self.students = []
    
    def accept_details(self):
        name = input("Enter your name : ")
        roll_no = int(input('Enter your roll_no : '))
        social_mark = int(input("Enter your social_mark : "))
        science_mark = int(input("Enter your science_mark : "))
        student = Student_Details(name,roll_no,social_mark,science_mark)
        self.students.append(student)
        print("Added student details successfully")
        
    def display_details(self):
        
        if not self.students:
            print("No student details found")
            return
        for student in self.students:
            student.display_details()
            
    def search_details(self):
        
        roll_no = int(input("Enter the Roll_no to search: "))
        found = False
        for student in self.students:
            if student.roll_no == roll_no:
                student.display_details()
                found = True
                break
        if not found:
            print("No student details is found {}".format(roll_no))
            
    def delete_details(self):
        
        roll_no = int(input("Enter the roll_no to delete: "))
        
        for student in self.students:
            if student.roll_no == roll_no:
                self.students.remove(student)
                print("Student with this roll no {} is deleted successfully".format(roll_no))
                return
            else:
                print("No students details are found for this {} roll_no".format(roll_no))
            
        
    def update_details(self):
        
        old_rollno = int(input("Enter the old roll no"))
        
        New_rollno = int(input("Enter the new roll no"))
        
        for student in self.students:
            if student.roll_no == old_rollno:
                student.roll_no = New_rollno
                print('The roll number has been updated succefully')
                return
        print('No student details found for this {} Roll_no'.format(old_rollno))
        
def main():
    
    sms = StudentManagement()
    
    while True:
        print("Menu: ")
        print("1. Add Student details")
        print("2. Display Student details")
        print("3. Search Student details")
        print("4. Delete Student details")
        print("5. Update Student details")
        print("6. Exit")
        number=int(input("Please enter the number in the menu"))
        if number == 1:
            sms.accept_details()
        elif number == 2:
            sms.display_details()
        elif number == 3:
            sms.search_details()
        elif number == 4:
            sms.delete_details()
        elif number == 5:
            sms.update_details()
        elif number == 6:
            break
        else:
            print("Invalid number please choose the correct number from the menu")

if __name__ == '__main__':
    main()
                           
                        
```
